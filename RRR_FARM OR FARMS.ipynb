```python
import pandas as pd 
import numpy as np
import re 
```

```python
data  = pd.read_csv('D:\SBA\EIDL/DATAACT_EIDL_LOANS_20200724-20201115.csv')
data.head() 
```

```python
data.columns 
```

```python
data.isnull().sum()
```

```python
The feature 'AWARDEEORRECIPIENTLEGALENTITYNAME' has 189 missing values, so we will replace them as with 'xxx'
```

```python
data['AWARDEEORRECIPIENTLEGALENTITYNAME'] = data['AWARDEEORRECIPIENTLEGALENTITYNAME'].replace(np.nan, 'XXX')
```

```python
g =[]
for name in data['AWARDEEORRECIPIENTLEGALENTITYNAME']: #iterating through every row
    match = re.search(r'\bFARMS\b',name) or re.search(r'\bFARM\b',name) # search for 'FARMS' or 'FARM' in every sentence
    if match:
         g.append(1)
    else:
        g.append(0)
''' if the a sentence contains FARM OR FARMS it will append 1 to the list (g), and append 0 to g'''
```

```python
data['Results'] = g 
dff = data[['FAIN','AWARDEEORRECIPIENTLEGALENTITYNAME','Results']] 
dff.head()
```
